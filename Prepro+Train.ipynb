```python
# json to csv + 문장 단위 구분 (^)
import json
import csv

with open('/content/drive/MyDrive/Colab_Notebooks/Study/DACON/train_original.json', 'r', encoding = 'utf-8-sig') as input_file, open('/content/drive/MyDrive/Colab_Notebooks/Study/DACON/train_original.csv', 'w', newline = '', encoding = 'utf-8-sig') as output_file:
    json_data = json.load(input_file)

    f = csv.writer(output_file)
    f.writerow(["article_original", "extractive"])
        
    for article in range(len(json_data["documents"])): 
        art = []
        for paragraph in range(len(json_data["documents"][article]["text"])):
            for sentence in range(len(json_data["documents"][article]["text"][paragraph])):
                json_article_original = json_data["documents"][article]["text"][paragraph][sentence]["sentence"]
                art.append(json_article_original)
        json_extract = json_data["documents"][article]["extractive"] 

        art_join = "^".join(art)
        art_join_list = []
        art_join_list.append(art_join)
        f.writerow([art_join_list, json_extract])
```

## Load CSV

```python
# load 후 확인해보기
import pandas as pd
import json

test = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/Study/BertSum-master/output.csv', encoding='utf-8-sig').drop('Unnamed: 0', axis=1)
test.tail()
```

```python
import ast
from tqdm import tqdm
import os

list_dic = []
for idx, row in test.iterrows():
  raw = row['morphs']
  target_idx = ast.literal_eval(row['extractive'])

  sentences = raw.split('^')
  src = [i.split(' ') for i in sentences]
  tgt = [a for i,a in enumerate(src) if i in target_idx]
  
  mydict = {}
  mydict['src'] = src
  mydict['tgt'] = tgt
  list_dic.append(mydict)
        
temp = []
for i,a in enumerate(tqdm(list_dic)):
  if (i+1)%6!=0:
      temp.append(a)
  else:
      filename = 'korean.'+'train'+'.'+str(i//6)+'.json'
      with open('/content/drive/MyDrive/Colab_Notebooks/Study/BertSum-master'+'/json_/'+filename, "w", newline = '', encoding = 'utf-8' ) as json_file:
          json.dump(temp, json_file, ensure_ascii=False)
      temp = []
```
